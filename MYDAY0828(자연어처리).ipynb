```python
(xtrain, ytrain), (xtest, ytest)=reuters.load_data(num_words=1000)
```

```python
max_len=100
```

```python
from keras.preprocessing import sequence
xtrain=sequence.pad_sequences(xtrain, maxlen=max_len)
xtest=sequence.pad_sequences(xtest, maxlen=max_len)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding
from keras.utils import np_utils
```

```python
ytrain=np_utils.to_categorical(ytrain)
ytest=np_utils.to_categorical(ytest)
```

```python
model=Sequential()
model.add(Embedding(1000,128))
```

```python
model.add(LSTM(128))
model.add(Dense(46, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(xtrain, ytrain, batch_size=100, epochs= 20, validation_data=(xtest, ytest))
```

```python
import re
```

```python
trainData['document']=trainData['document'].str.replace("[^ㄱ-ㅎ ㅏ-ㅣ가-힣]","")
trainData[:5]
```

```python
import konlpy
from konlpy.tag import Okt
okt=Okt()
okt.morphs("흠포스터보고 초딩영화줄오버연기조차 가볍지 않구나")
```

```python
stopwords=['의', '가', '이', '은','하다','한','와', '에','으로', '자', '과','도', '걍',
           '잘','좀', '는','들','을','를','것', '나']
```

```python
okt.morphs('에이 이런 것도 영화냐? 차라리 집에서 티브이나 보겠다', stem=True)
```

```python
xtrain=[]
for sentence in trainData['document']:
    tempx=[]
    tempx=okt.morphs(sentence, stem=True)
    tempx=[word for word in tempx if not word in stopwords]
    xtrain.append(tempx)
        
```

```python
testData=testData.dropna(how='any')
testData[testData.document.isnull()]

testData['document']=testData['document'].str.replace("[^ㄱ-ㅎ ㅏ-ㅣ가-힣]","")

xtest=[]
for sentence in testData['document']:
    tempx=[]
    tempx=okt.morphs(sentence, stem=True)
    tempx=[word for word in tempx if not word in stopwords]
    xtest.append(tempx)
        
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(num_words=30000)
tokenizer.fit_on_texts(xtrain)
xtrain=tokenizer.texts_to_sequences(xtrain)
xtest=tokenizer.texts_to_sequences(xtest)
```

```python
ytrain=trainData['label']
ytest=testData['label']
```

```python
from keras.layers import *
from keras.models import Sequential
from keras.preprocessing.sequence import pad_sequences
```

```python
maxLen=30
xtrain=pad_sequences(xtrain,maxlen=maxLen)
xtest=pad_sequences(xtest, maxlen=maxLen)
```

```python
model=Sequential()
model.add(Embedding(30000,100))
model.add(LSTM(128))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
```

```python
history=model.fit(xtrain, ytrain, epochs=4, batch_size=60, validation_split=0.2)
```
