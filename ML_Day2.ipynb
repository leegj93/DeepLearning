```python
#decision tree
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("data/titanic/train.csv")
test=pd.read_csv("data/titanic/test.csv")
```

```python
train.describe()
train.isnull().sum()
print(train['Age'].mean())
```

```python
train['Age_mean']=train['Age']
train['Age_mean'].fillna(train['Age_mean'].mean(),inplace=True)
test['Age_mean']=test['Age']
test['Age_mean'].fillna(test['Age_mean'].mean(),inplace=True)
```

```python
train['Sex'].value_counts()
```

```python
train['Gender']=(train['Sex']=='female')
test['Gender']=(test['Sex']=='female')
```

```python
train['Embarked'].value_counts()
train['Embarked'].isnull().sum() #2
```

```python
print(test['Embarked'].isnull().sum())
```

```python
train['Embarked_S']=train['Embarked']=='S'
train['Embarked_C']=train['Embarked']=='C'
train['Embarked_Q']=train['Embarked']=='Q'
#S:100, C:010, Q:001
```

```python
train['Embarked'].value_counts()
train['Embarked_S'].sum()
```

```python
train[['Embarked_S','Embarked_C','Embarked_Q']]
```

```python
test['Embarked_S']=test['Embarked']=='S'
test['Embarked_C']=test['Embarked']=='C'
test['Embarked_Q']=test['Embarked']=='Q'
```

```python
train['FamilySize']=train['SibSp']+train['Parch']+1
train['FamilySize'].value_counts()
train['Family']=train['FamilySize']
#대가족(5이상):L, 소가족(1):S, 중가족(2~4):M

train.loc[train['FamilySize']==1,'Family']='S'
train.loc[(train['FamilySize']>1) & (train['FamilySize']<5),'Family']='M'
train.loc[train['FamilySize']>4,'Family']='L'
train[['Family', 'FamilySize']]
```

```python
train['Family_S']=train['Family']=='S'
train['Family_M']=train['Family']=='M'
train['Family_L']=train['Family']=='L'
```

```python
train[['Family_S', 'Family_M','Family_L','Family',
       'FamilySize']]
```

```python
test['FamilySize']=test['SibSp']+test['Parch']+1
test['FamilySize'].value_counts()
test['Family']=test['FamilySize']
#대가족(5이상):L, 소가족(1):S, 중가족(2~4):M

test.loc[train['FamilySize']==1,'Family']='S'
test.loc[(train['FamilySize']>1) & (test['FamilySize']<5),'Family']='M'
test.loc[train['FamilySize']>4,'Family']='L'
test[['Family', 'FamilySize']]
```

```python
test['Family_S']=test['Family']=='S'
test['Family_M']=test['Family']=='M'
test['Family_L']=test['Family']=='L'
```

```python
test[['Family_S', 'Family_M','Family_L','Family',
       'FamilySize']]
```

```python
train['Pclass'].dtypes
```

```python
train['Pclass']=train['Pclass'].astype('category')
train['Pclass'].dtypes
```

```python
fn=['Gender', 'Age_mean', 'Embarked_S', 'Embarked_C', 
    'Embarked_Q', 'Family_S', 'Family_M', 'Family_L']
#train.info()
X_train=train[fn]
X_train.head()
y_label=train['Survived']
X_test=test[fn]
```

```python
#모델링(디시전 트리)
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_depth=3)
model
```

```python
model.fit(X_train, y_label)
```

```python
from sklearn.model_selection import KFold, cross_val_score
kfold=KFold(n_splits=10, shuffle=True, random_state=2019)#n_splits=3, random_state=None, shuffle=False
kfold
```

```python
score=cross_val_score(model, X_train, y_label, 
                      cv=kfold, scoring='accuracy')
print(score)
np.mean(score)
```

```python
prediction=model.predict(X_test)
prediction.shape
```

```python
test['Survived']=prediction
test.columns
```

```python
submission=test[['PassengerId','Survived']]
submission.to_csv("submit.csv", index=False)
```
